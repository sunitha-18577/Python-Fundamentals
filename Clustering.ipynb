```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans,DBSCAN
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import make_blobs
```

```python
x,y=make_blobs(n_samples=300,centers=4,random_state=42,cluster_std=0.60)
scaler=StandardScaler()
x_scaled=scaler.fit_transform(x)
```

```python
kmeans=KMeans(n_clusters=4,random_state=42,n_init=10)
kmeans_labels=kmeans.fit_predict(x_scaled)
```

```python
from sklearn.cluster import AgglomerativeClustering
hierarchical=AgglomerativeClustering(n_clusters=4)
hierarchical_labels=hierarchical.fit_predict(x_scaled)
```

```python
dbscan=DBSCAN(eps=0.5,min_samples=5)
dbscan_labels=dbscan.fit_predict(x_scaled)
```

```python
plt.figure(figsize=(15,5))
```

```python
plt.subplot(1,3,1)
plt.scatter(x_scaled[:,0],x_scaled[:,1],c=kmeans_labels,cmap='viridis')
plt.title("K-Means")
```

```python
plt.subplot(1,3,2)
plt.scatter(x_scaled[:,0],x_scaled[:,1],c=hierarchical_labels,cmap='viridis')
plt.title("Hierarchical")
```

```python
plt.subplot(1,3,3)
plt.scatter(x_scaled[:,0],x_scaled[:,1],c=dbscan_labels,cmap='viridis')
plt.title("DBSCAN")
```

```python
plt.show()
```

```python
kmeans_silhouette=silhouette_score(x_scaled,kmeans_labels)
dbscan_silhouette=silhouette_score(x_scaled,dbscan_labels)
```

```python
print(f'K-Means Silhouette Score: {kmeans_silhouette}')
print(f'DBSCAN Silhouette Score: {dbscan_silhouette}')
```

```python
scaler=StandardScaler()
x_scaled=scaler.fit_transform(x)
```

```python
kmeans=KMeans(n_clusters=4,random_state=42,n_init=10)
kmeans_labels=kmeans.fit_predict(x_scaled)
kmeans_silhouette=silhouette_score(x_scaled,kmeans_labels)
```

```python
hierarchical=AgglomerativeClustering(n_clusters=4)
hierarchical_labels=hierarchical.fit_predict(x_scaled)
hierarchical_silhouette=silhouette_score(x_scaled,hierarchical_labels)
```

```python
dbscan=DBSCAN(eps=0.5,min_samples=5)
dbscan_labels=dbscan.fit_predict(x_scaled)
dbscan_silhouette=silhouette_score(x_scaled,dbscan_labels)
```

```python
dbscan_labels_no_noise=dbscan_labels[dbscan_labels!=-1]
x_scaled_no_noise=x_scaled[dbscan_labels!=-1]
```

```python
if len(dbscan_labels_no_noise)>0:
    dbscan_silhouette_no_noise=silhouette_score(x_scaled_no_noise,dbscan_labels_no_noise)
else:
    dbscan_silhouette_no_noise=0
```

```python
plt.figure(figsize=(15,5))
```

```python
plt.subplot(1,3,1)
plt.scatter(x_scaled[:,0],x_scaled[:,1],c=kmeans_labels,cmap='viridis',marker='o',edgecolors='k')
plt.title(f'K-Means (Silhouette Score: {kmeans_silhouette:.2f})')
```

```python
plt.subplot(1,3,2)
plt.scatter(x_scaled[:,0],x_scaled[:,1],c=hierarchical_labels,cmap='viridis',marker='o',edgecolors='k')
plt.title(f'Hierarchical (Silhouette Score: {hierarchical_silhouette:.2f})')
```

```python
plt.subplot(1,3,3)
plt.scatter(x_scaled[:,0],x_scaled[:,1],c=dbscan_labels,cmap='viridis',marker='o',edgecolors='k')
plt.title(f'DBSCAN (Silhouette Score: {dbscan_silhouette:.2f})')
```

```python
print("\nConclusions:")
print("The Silhoutte score provides a measure of how well each data point fits within the cluster")
print(f"score closer to 1 indicate better-defined clusters.Based on the scores")
print(f"- K-Means Silhouette Score: {kmeans_silhouette:.2f}")
print(f"- Hierarchical Silhouette Score: {hierarchical_silhouette:.2f}")
print(f"- DBSCAN Silhouette Score: {dbscan_silhouette:.2f}")
```
